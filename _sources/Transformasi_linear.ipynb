### 1. Rotasi

Rotasi adalah transformasi yang memutar objek sekitar titik tertentu. Untuk rotasi dua dimensi, kita menggunakan matriks rotasi berikut:

\[
\begin{bmatrix}
\cos(\theta) & -\sin(\theta) \\
\sin(\theta) & \cos(\theta)
\end{bmatrix}
\]

di mana \(\theta\) adalah sudut rotasi dalam radian.

**Contoh Penyelesaian**:
Misalkan kita ingin memutar sebuah titik \(P(1, 1)\) sebesar 45 derajat berlawanan arah jarum jam.

1. Hitung nilai \(\cos(45^\circ)\) dan \(\sin(45^\circ)\):
   \[
   \cos(45^\circ) = \frac{\sqrt{2}}{2}, \quad \sin(45^\circ) = \frac{\sqrt{2}}{2}
   \]

2. Substitusikan nilai-nilai tersebut ke dalam matriks rotasi:
   \[
   \begin{bmatrix}
   \frac{\sqrt{2}}{2} & -\frac{\sqrt{2}}{2} \\
   \frac{\sqrt{2}}{2} & \frac{\sqrt{2}}{2}
   \end{bmatrix}
   \]

### 2. Penskalaan

Penskalaan adalah transformasi yang mengubah ukuran objek. Untuk penskalaan dua dimensi, kita menggunakan matriks penskalaan berikut:

\[
\begin{bmatrix}
s_x & 0 \\
0 & s_y
\end{bmatrix}
\]

di mana \(s_x\) adalah faktor penskalaan pada sumbu x dan \(s_y\) adalah faktor penskalaan pada sumbu y.

**Contoh Penyelesaian**:
Misalkan kita ingin memperbesar sebuah segitiga dengan faktor skala 2 pada sumbu x dan 3 pada sumbu y.

1. Buat matriks penskalaan:
   \[
   \begin{bmatrix}
   2 & 0 \\
   0 & 3
   \end{bmatrix}
   \]

### 3. Translasi

Translasi adalah transformasi yang memindahkan objek. Untuk translasi dua dimensi, kita menggunakan matriks translasi berikut:

\[
\begin{bmatrix}
1 & 0 & t_x \\
0 & 1 & t_y \\
0 & 0 & 1
\end{bmatrix}
\]

di mana \(t_x\) adalah jumlah pergeseran pada sumbu x dan \(t_y\) adalah jumlah pergeseran pada sumbu y.

**Contoh Penyelesaian**:
Misalkan kita ingin memindahkan sebuah kotak sejajar sumbu x sebesar 3 satuan dan sejajar sumbu y sebesar -2 satuan.

1. Buat matriks translasi:
   \[
   \begin{bmatrix}
   1 & 0 & 3 \\
   0 & 1 & -2 \\
   0 & 0 & 1
   \end{bmatrix}
   \]

### 4. Refleksi

Refleksi adalah transformasi yang mencerminkan objek terhadap sebuah garis. Untuk refleksi dua dimensi, kita menggunakan matriks refleksi berikut:

- Refleksi terhadap sumbu x:
\[
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}
\]

- Refleksi terhadap sumbu y:
\[
\begin{bmatrix}
-1 & 0 \\
0 & 1
\end{bmatrix}
\]

**Contoh Penyelesaian**:
Misalkan kita ingin merefleksikan sebuah lingkaran terhadap sumbu x.

1. Buat matriks refleksi:
   \[
   \begin{bmatrix}
   1 & 0 \\
   0 & -1
   \end{bmatrix}
   \]

Ini adalah beberapa contoh jenis matriks transformasi berserta rumus dan penyelesaiannya. Matriks-matriks ini membantu untuk mentransformasikan objek dalam ruang dua dimensi.

# Konsep Rotasi pada Matriks Transformasi

Rotasi adalah transformasi yang mengubah posisi suatu objek dengan memutarnya sekitar titik tertentu, biasanya titik asal (0, 0) dalam sistem koordinat. Rotasi dua dimensi dapat direpresentasikan menggunakan matriks rotasi.

## Matriks Rotasi

Untuk rotasi dengan sudut \(\theta\) berlawanan arah jarum jam, matriks rotasi yang digunakan adalah:

\[
R(\theta) = \begin{bmatrix}
\cos(\theta) & -\sin(\theta) \\
\sin(\theta) & \cos(\theta)
\end{bmatrix}
\]

## Penjelasan Elemen Matriks

- \(\cos(\theta)\) dan \(\sin(\theta)\) adalah fungsi trigonometri yang menentukan perubahan posisi titik setelah rotasi.
- Elemen \(-\sin(\theta)\) dan \(\cos(\theta)\) mengatur orientasi titik dalam rotasi dua dimensi.

## Cara Kerja Rotasi

Misalkan kita memiliki titik \(P(x, y)\) dan ingin memutar titik tersebut sebesar sudut \(\theta\) berlawanan arah jarum jam. Koordinat baru \(P'(x', y')\) setelah rotasi adalah hasil perkalian antara matriks rotasi \(R(\theta)\) dengan vektor koordinat awal \(\begin{bmatrix} x \\ y \end{bmatrix}\):

\[
\begin{bmatrix}
x' \\
y'
\end{bmatrix} = \begin{bmatrix}
\cos(\theta) & -\sin(\theta) \\
\sin(\theta) & \cos(\theta)
\end{bmatrix} \begin{bmatrix}
x \\
y
\end{bmatrix}
\]

## Contoh Penerapan

Misalkan kita ingin memutar titik \(P(1, 0)\) sebesar 90 derajat berlawanan arah jarum jam.

1. Sudut rotasi: \(\theta = 90^\circ\) atau \(\theta = \frac{\pi}{2}\) radian.
2. Matriks rotasi untuk 90 derajat:
   \[
   R\left(\frac{\pi}{2}\right) = \begin{bmatrix}
   \cos\left(\frac{\pi}{2}\right) & -\sin\left(\frac{\pi}{2}\right) \\
   \sin\left(\frac{\pi}{2}\right) & \cos\left(\frac{\pi}{2}\right)
   \end{bmatrix} = \begin{bmatrix}
   0 & -1 \\
   1 & 0
   \end{bmatrix}
   \]

3. Koordinat baru setelah rotasi:
   \[
   \begin{bmatrix}
   x' \\
   y'
   \end{bmatrix} = \begin{bmatrix}
   0 & -1 \\
   1 & 0
   \end{bmatrix} \begin{bmatrix}
   1 \\
   0
   \end{bmatrix} = \begin{bmatrix}
   0 \cdot 1 + (-1) \cdot 0 \\
   1 \cdot 1 + 0 \cdot 0
   \end{bmatrix} = \begin{bmatrix}
   0 \\
   1
   \end{bmatrix}
   \]

Jadi, titik \(P(1, 0)\) setelah diputar 90 derajat menjadi titik \(P'(0, 1)\).

```python
#Contoh Kode
import numpy as np
import matplotlib.pyplot as plt

def apply_transformation(matrix, transformation_matrix):
    return np.dot(transformation_matrix, matrix)

def plot_transformation(original_points, transformed_points, title):
    plt.figure()
    plt.plot(original_points[0, :], original_points[1, :], 'bo-', label='Original')
    plt.plot(transformed_points[0, :], transformed_points[1, :], 'ro-', label='Transformed')
    plt.axhline(0, color='grey', lw=0.5)
    plt.axvline(0, color='grey', lw=0.5)
    plt.grid(True)
    plt.legend()
    plt.title(title)
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

# Titik-titik koordinat yang akan ditransformasikan
points = np.array([[1, 2, -1, -2],
                   [1, -1, 2, -2]])

# Sudut rotasi dalam radian
theta = np.pi / 4  # 45 derajat

# Matriks rotasi
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])

# Transformasi Rotasi
transformed_points = apply_transformation(points, rotation_matrix)
plot_transformation(points, transformed_points, "Rotation 45¬∞")
```

# Konsep Penskalaan pada Matriks Transformasi

Penskalaan (scaling) adalah transformasi yang mengubah ukuran suatu objek dalam suatu ruang tanpa mengubah bentuknya. Penskalaan dilakukan dengan memperbesar atau memperkecil koordinat titik-titik dari objek tersebut.

## Matriks Penskalaan

Matriks penskalaan dua dimensi memiliki bentuk sebagai berikut:

\[
S = \begin{bmatrix}
s_x & 0 \\
0 & s_y
\end{bmatrix}
\]

- \(s_x\) adalah faktor skala dalam arah sumbu x.
- \(s_y\) adalah faktor skala dalam arah sumbu y.

## Penjelasan Elemen Matriks

- Elemen \(s_x\) memperbesar atau memperkecil koordinat x dari titik-titik objek.
- Elemen \(s_y\) memperbesar atau memperkecil koordinat y dari titik-titik objek.

## Cara Kerja Penskalaan

Misalkan kita memiliki titik \(P(x, y)\) dan ingin menskalakannya dengan faktor \(s_x\) untuk sumbu x dan \(s_y\) untuk sumbu y. Koordinat baru \(P'(x', y')\) setelah penskalaan adalah hasil perkalian antara matriks penskalaan \(S\) dengan vektor koordinat awal \(\begin{bmatrix} x \\ y \end{bmatrix}\):

\[
\begin{bmatrix}
x' \\
y'
\end{bmatrix} = \begin{bmatrix}
s_x & 0 \\
0 & s_y
\end{bmatrix} \begin{bmatrix}
x \\
y
\end{bmatrix} = \begin{bmatrix}
s_x \cdot x \\
s_y \cdot y
\end{bmatrix}
\]

## Contoh Penerapan

Misalkan kita ingin menskalakan titik \(P(1, 2)\) dengan faktor skala 2 untuk sumbu x dan faktor skala 3 untuk sumbu y.

1. Faktor skala: \(s_x = 2\), \(s_y = 3\).
2. Matriks penskalaan:
   \[
   S = \begin{bmatrix}
   2 & 0 \\
   0 & 3
   \end{bmatrix}
   \]

3. Koordinat baru setelah penskalaan:
   \[
   \begin{bmatrix}
   x' \\
   y'
   \end{bmatrix} = \begin{bmatrix}
   2 & 0 \\
   0 & 3
   \end{bmatrix} \begin{bmatrix}
   1 \\
   2
   \end{bmatrix} = \begin{bmatrix}
   2 \cdot 1 \\
   3 \cdot 2
   \end{bmatrix} = \begin{bmatrix}
   2 \\
   6
   \end{bmatrix}
   \]

Jadi, titik \(P(1, 2)\) setelah penskalaan menjadi titik \(P'(2, 6)\).

```python
# Contoh Kode
import numpy as np
import matplotlib.pyplot as plt

def apply_transformation(matrix, transformation_matrix):
    return np.dot(transformation_matrix, matrix)

def plot_transformation(original_points, transformed_points, title):
    plt.figure()
    plt.plot(original_points[0, :], original_points[1, :], 'bo-', label='Original')
    plt.plot(transformed_points[0, :], transformed_points[1, :], 'ro-', label='Transformed')
    plt.axhline(0, color='grey', lw=0.5)
    plt.axvline(0, color='grey', lw=0.5)
    plt.grid(True)
    plt.legend()
    plt.title(title)
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

# Titik-titik koordinat yang akan ditransformasikan
points = np.array([[1, 2, -1, -2],
                   [1, -1, 2, -2]])

# Matriks penskalaan
scaling_matrix = np.array([[2, 0],
                           [0, 1]])

# Transformasi Penskalaan
transformed_points = apply_transformation(points, scaling_matrix)
plot_transformation(points, transformed_points, "Scaling 2x in X direction")
```

## Penjelasan Langkah-langkah Kode

 - apply_transformation(matrix, transformation_matrix) melakukan perkalian matriks antara matriks titik-titik dan matriks transformasi.
 - plot_transformation(original_points, transformed_points, title) memplot titik-titik asli dan titik-titik setelah transformasi.

2. Titik-titik Koordinat:

 - Titik-titik koordinat yang akan ditransformasikan didefinisikan sebagai points.

3. Matriks Penskalaan:

 - Matriks penskalaan scaling_matrix didefinisikan dengan faktor skala 2 pada sumbu x dan 1 pada sumbu y. Ini berarti titik-titik pada sumbu x akan diperbesar dua kali lipat, sedangkan titik-titik pada sumbu y tidak akan berubah.

4. Penerapan Transformasi:

 - Transformasi penskalaan diterapkan pada titik-titik koordinat menggunakan apply_transformation.
 - Hasil transformasi diplot menggunakan plot_transformation.

Contoh Hasil Transformasi

Misalnya, titik (1, 1) setelah penskalaan dengan faktor (2, 1) akan menjadi (2, 1), titik (2, -1) akan menjadi (4, -1), dan seterusnya. Plot hasilnya akan menunjukkan titik-titik asli (berwarna biru) dan titik-titik setelah penskalaan (berwarna merah). Titik-titik hasil penskalaan akan lebih jauh dari sumbu y dibandingkan titik-titik asli karena diperbesar dua kali lipat di arah x.

# Konsep Translasi pada Matriks Transformasi

Translasi adalah salah satu jenis transformasi geometri yang memindahkan setiap titik dari suatu objek dengan jarak yang sama dalam arah yang sama. Ini berarti bahwa setiap titik pada objek akan berpindah ke lokasi baru, tetapi objek itu sendiri tidak akan berubah bentuk atau ukurannya.

## Matriks Translasi

Berbeda dengan transformasi linear seperti rotasi, penskalaan, dan refleksi, translasi tidak bisa diwakili langsung oleh matriks \(2 \times 2\). Sebaliknya, kita menggunakan representasi matriks homogen \(3 \times 3\) untuk melakukan translasi.

Matriks translasi dalam bentuk homogen dua dimensi adalah sebagai berikut:

\[
T = \begin{bmatrix}
1 & 0 & t_x \\
0 & 1 & t_y \\
0 & 0 & 1
\end{bmatrix}
\]

- \(t_x\) adalah jarak translasi dalam arah sumbu x.
- \(t_y\) adalah jarak translasi dalam arah sumbu y.

## Cara Kerja Translasi

Untuk menerapkan translasi pada titik \(P(x, y)\), kita perlu menggunakan koordinat homogen dengan menambahkan komponen \(1\) sebagai komponen ketiga, sehingga vektor titik menjadi \(\begin{bmatrix} x \\ y \\ 1 \end{bmatrix}\).

Koordinat baru \(P'(x', y')\) setelah translasi diberikan oleh:

\[
\begin{bmatrix}
x' \\
y' \\
1
\end{bmatrix} = \begin{bmatrix}
1 & 0 & t_x \\
0 & 1 & t_y \\
0 & 0 & 1
\end{bmatrix} \begin{bmatrix}
x \\
y \\
1
\end{bmatrix} = \begin{bmatrix}
x + t_x \\
y + t_y \\
1
\end{bmatrix}
\]

## Contoh Penerapan

Misalkan kita ingin mentranslasi titik \(P(1, 2)\) dengan jarak translasi \(t_x = 3\) dan \(t_y = 4\).

1. Matriks translasi:
   \[
   T = \begin{bmatrix}
   1 & 0 & 3 \\
   0 & 1 & 4 \\
   0 & 0 & 1
   \end{bmatrix}
   \]

2. Koordinat homogen dari titik \(P\):
   \[
   \begin{bmatrix}
   1 \\
   2 \\
   1
   \end{bmatrix}
   \]

3. Koordinat baru setelah translasi:
   \[
   \begin{bmatrix}
   x' \\
   y' \\
   1
   \end{bmatrix} = \begin{bmatrix}
   1 & 0 & 3 \\
   0 & 1 & 4 \\
   0 & 0 & 1
   \end{bmatrix} \begin{bmatrix}
   1 \\
   2 \\
   1
   \end{bmatrix} = \begin{bmatrix}
   1 + 3 \\
   2 + 4 \\
   1
   \end{bmatrix} = \begin{bmatrix}
   4 \\
   6 \\
   1
   \end{bmatrix}
   \]

Jadi, titik \(P(1, 2)\) setelah translasi menjadi titik \(P'(4, 6)\).


```python
#Contoh Code
import numpy as np
import matplotlib.pyplot as plt

def apply_transformation_homogeneous(matrix, transformation_matrix):
    ones_row = np.ones((1, matrix.shape[1]))
    homogeneous_matrix = np.vstack([matrix, ones_row])
    transformed_homogeneous_matrix = np.dot(transformation_matrix, homogeneous_matrix)
    return transformed_homogeneous_matrix[:2, :]

def plot_transformation(original_points, transformed_points, title):
    plt.figure()
    plt.plot(original_points[0, :], original_points[1, :], 'bo-', label='Original')
    plt.plot(transformed_points[0, :], transformed_points[1, :], 'ro-', label='Transformed')
    plt.axhline(0, color='grey', lw=0.5)
    plt.axvline(0, color='grey', lw=0.5)
    plt.grid(True)
    plt.legend()
    plt.title(title)
    plt.xlim(-5, 10)
    plt.ylim(-5, 10)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

# Titik-titik koordinat yang akan ditransformasikan
points = np.array([[1, 2, -1, -2],
                   [1, -1, 2, -2]])

# Matriks translasi
translation_matrix = np.array([[1, 0, 3],
                               [0, 1, 4],
                               [0, 0, 1]])

# Transformasi Translasi
transformed_points = apply_transformation_homogeneous(points, translation_matrix)
plot_transformation(points, transformed_points, "Translation by (3, 4)")
```

##Penjelasan Langkah-langkah Kode
1. Import dan Definisi Fungsi:

 - apply_transformation_homogeneous(matrix, transformation_matrix) menambahkan baris 1 ke matriks titik untuk membuatnya menjadi koordinat homogen, menerapkan matriks transformasi homogen, dan kemudian menghapus baris homogen untuk kembali ke koordinat kartesian.
 - plot_transformation(original_points, transformed_points, title) memplot titik-titik asli dan hasil transformasi.

2. Titik-titik Koordinat:

 - Titik-titik koordinat yang akan ditransformasikan didefinisikan sebagai points.

3. Matriks Translasi:

 - Matriks translasi translation_matrix didefinisikan untuk menggeser titik-titik sebesar (3, 4).

4. Penerapan Transformasi:

 - Transformasi translasi diterapkan pada titik-titik koordinat menggunakan apply_transformation_homogeneous.
Hasil transformasi diplot menggunakan plot_transformation.

Contoh Hasil Transformasi

Misalnya, titik (1, 1) setelah ditranslasi dengan vektor (3, 4) akan menjadi (4, 5), dan seterusnya untuk titik-titik lainnya. Plot hasilnya akan menunjukkan titik-titik asli dan titik-titik setelah ditranslasi, dengan titik-titik baru digeser sesuai dengan vektor translasi yang diberikan.

# Konsep Reflection (Refleksi) pada Matriks Transformasi

Refleksi adalah salah satu jenis transformasi geometri yang mencerminkan setiap titik dari suatu objek melalui suatu garis (dalam dua dimensi) atau bidang (dalam tiga dimensi). Refleksi mengubah orientasi objek tetapi mempertahankan bentuk dan ukurannya.

## Matriks Refleksi dalam Dua Dimensi

Refleksi dapat diwakili oleh matriks transformasi \(2 \times 2\) atau matriks homogen \(3 \times 3\). Berikut adalah beberapa contoh matriks refleksi dalam dua dimensi:

1. **Refleksi terhadap sumbu x:**
   \[
   R_x = \begin{bmatrix}
   1 & 0 \\
   0 & -1
   \end{bmatrix}
   \]
   Matriks ini mencerminkan titik di sekitar sumbu x.

2. **Refleksi terhadap sumbu y:**
   \[
   R_y = \begin{bmatrix}
   -1 & 0 \\
   0 & 1
   \end{bmatrix}
   \]
   Matriks ini mencerminkan titik di sekitar sumbu y.

3. **Refleksi terhadap garis y = x:**
   \[
   R_{y=x} = \begin{bmatrix}
   0 & 1 \\
   1 & 0
   \end{bmatrix}
   \]
   Matriks ini mencerminkan titik di sekitar garis \(y = x\).

4. **Refleksi terhadap garis y = -x:**
   \[
   R_{y=-x} = \begin{bmatrix}
   0 & -1 \\
   -1 & 0
   \end{bmatrix}
   \]
   Matriks ini mencerminkan titik di sekitar garis \(y = -x\).

## Cara Kerja Refleksi

Untuk menerapkan refleksi pada titik \(P(x, y)\), kita mengalikan vektor koordinat titik dengan matriks refleksi yang sesuai. Koordinat baru \(P'(x', y')\) setelah refleksi diberikan oleh:

\[
\begin{bmatrix}
x' \\
y'
\end{bmatrix} = R \begin{bmatrix}
x \\
y
\end{bmatrix}
\]

## Contoh Penerapan

Misalkan kita ingin mencerminkan titik \(P(3, 4)\) terhadap sumbu x.

1. Matriks refleksi terhadap sumbu x:
   \[
   R_x = \begin{bmatrix}
   1 & 0 \\
   0 & -1
   \end{bmatrix}
   \]

2. Koordinat titik \(P\):
   \[
   \begin{bmatrix}
   3 \\
   4
   \end{bmatrix}
   \]

3. Koordinat baru setelah refleksi:
   \[
   \begin{bmatrix}
   x' \\
   y'
   \end{bmatrix} = \begin{bmatrix}
   1 & 0 \\
   0 & -1
   \end{bmatrix} \begin{bmatrix}
   3 \\
   4
   \end{bmatrix} = \begin{bmatrix}
   3 \\
   -4
   \end{bmatrix}
   \]

Jadi, titik \(P(3, 4)\) setelah refleksi terhadap sumbu x menjadi titik \(P'(3, -4)\).

```python
import numpy as np

def horizontal_reflection(matrix):
    return np.flipud(matrix)

def vertical_reflection(matrix):
    return np.fliplr(matrix)

def diagonal_reflection(matrix):
    return np.transpose(matrix)

# Contoh penggunaan
matrix = np.array([[1, 2],
                   [4, 5]])

print("Original Matrix:")
print(matrix)

print("\nHorizontal Reflection:")
print(horizontal_reflection(matrix))

print("\nVertical Reflection:")
print(vertical_reflection(matrix))

print("\nDiagonal Reflection:")
print(diagonal_reflection(matrix))
```

1. **`horizontal_reflection`**: Fungsi ini melakukan refleksi horizontal terhadap matriks yang diberikan. Untuk mencapai itu, fungsi menggunakan fungsi `np.flipud()`, yang membalikkan urutan baris matriks, menghasilkan efek refleksi horizontal.

2. **`vertical_reflection`**: Fungsi ini melakukan refleksi vertikal terhadap matriks yang diberikan. Ini dilakukan dengan menggunakan fungsi `np.fliplr()`, yang membalikkan urutan kolom matriks, menghasilkan efek refleksi vertikal.

3. **`diagonal_reflection`**: Fungsi ini melakukan refleksi diagonal terhadap matriks yang diberikan. Untuk mencapai ini, fungsi menggunakan fungsi `np.transpose()`, yang menukar baris dengan kolom dalam matriks, menghasilkan efek refleksi diagonal.
