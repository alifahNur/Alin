```python
import numpy as np

# Matriks A
A = np.array([[10, 7, 9],
              [9, 8, 10],
              [8, 9, 11]])
print("\nMatriks A :")
print(A)

# Matriks B
B = np.array([[1, 2, 3],
              [0, 3, 2],
              [-1, 4, 1]])
print("\nMatriks B :")
print(B)
```

```python
#Penjumlahan

C = A + B
print("\nMatriks C :")
print(C)

D = B + C
print("\nMatriks D :")
print(D)

#Pengurangan

E = D - C
print("\nMatriks E : ")
print(E)

F = A - E
print("\nMatriks F :")
print(F)

#Transpose

A_T = A.transpose()
print('\nMatriks A Transpose')
print(A_T)

B_T = B.transpose()
print('\nMatriks B Transpose')
print(B_T)

C_T = C.transpose()
print('\nMatriks C Transpose')
print(C_T)

D_T = D.transpose()
print('\nMatriks D Transpose')
print(D_T)

E_T = A.transpose()
print('\nMatriks E Transpose')
print(E_T)

F_T = F.transpose()
print('\nMatriks F Transpose')
print(F_T)

#perkalian
G = A @ B
print ("\nMatriks G :")
print(G)

# Fungsi untuk mengalikan dua matriks
def multiply_matrices(A, B):
    result = np.zeros((A.shape[0], B.shape[1]))
    for i in range(A.shape[0]):
        for j in range(B.shape[1]):
            for k in range(A.shape[1]):
                result[i][j] += A[i][k] * B[k][j]
    return result


# Mengalikan dua matriks tanpa menggunakan @
result = multiply_matrices(A, B)

# Menampilkan hasil
print("\nHasil perkalian matriks A dan B:")
print(result)
```
