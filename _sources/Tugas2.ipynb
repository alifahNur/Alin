- Berikan penjelasan tentang apa itu matrix

  *matriks adalah kumpulan bilangan atau elemen yang disusun dalam bentuk baris dan kolom. Setiap elemen dalam matriks diidentifikasi oleh baris dan kolomnya.
  matriks digunakan untuk merepresentasikan data, sistem linear, transformasi linear, dan berbagai konsep matematika lainnya.*

- jenis jenis matriks

  1. matriks nol : matriks yang seluruh elemennya adalah bilangan nol.
  2. matriks baris : matriks yang hanya memiliki satu baris.
  3. matriks kolom : matriks yang hanya memiliki satu kolom.
  4. matriks persegi : matriks yang banyaknya baris sama dengan banyaknya kolom.
  5. matriks diagonal : matriks persegi yang semua elemennya nol, kecuali pada diagonal utamanya.
  6. matriks segitiga atas, matriks persegi yang semua elemen di bawah diagonal utamanya adalah nol.
  7. matriks segitiga bawah : matriks persegi yang semua elemen di atas diagonal utamanya adalah nol.
  8. matriks identitas : matriks persegi yang elemen pada diagonal utamanya adalah satu, sedangkan elemen lainnya adalah nol.

- Operasi Matrix
  Operasi matriks dapat dilakukan hanya jika memenuhi syarat dan ketentuannya. Operasi matriks sendiri meliputi : penjumlahan dan pengurangan dua matriks, perkalian matriks dengan bilangan skalar, perkalian dua matriks, dan transpose matriks.

- penjumlahan dan pengurangan
  syarat : jika terdapat dua matriks, misal matriks j dan k, yang memiliki ordo sama, maka elemen yang seletak bisa dijumlahkan dan dikurangkan.

- Perkalian skalar
  pada operasi perkalian skalar, jika diketahui J merupakan suatu matriks dan K merupakan bilangan real, maka hasil perkalian K dengan matriks J adalah matriks yang diperoleh dengan mengalikan setiap elemen J dengan K.

- Perkalian dua matriks
  syarat perkalian dua matriks yaitu, kedua matriks memiliki jumlah kolom yang sama antara matriks A dengan Matriks B. untuk mencari hasil kali kedua matriks yitu dengan mengalikan elemen pada baris matriks A dengan elemen pada kolom matriks B, kemudian hasil perkalian antara baris dan kolom tersebut dijumlahkan.

- transpose matriks
  Transpose dari sebuah matriks adalah operasi untuk memindahkan elemen-elemen matriks ke posisi yang berlawanan secara diagonal. Artinya, jika sebuah matriks memiliki dimensi \( m \times n \), maka matriks transposenya akan memiliki dimensi \( n \times m \), dengan baris-baris dari matriks asli menjadi kolom-kolom dari matriks transposenya dan sebaliknya.

  Misalkan \( A \) adalah sebuah matriks dengan elemen \( a_{ij} \), di mana \( i \) menunjukkan indeks baris dan \( j \) menunjukkan indeks kolom. Transpose dari matriks \( A \), ditandai dengan \( A^T \), memiliki elemen yang sama, tetapi ditukar antara baris dan kolom, sehingga jika \( A^T \) adalah transpose dari \( A \), maka:

  \[ (A^T)_{ij} = a_{ji} \]


- buat program untuk mengalikan matriks a dan b tanpa menggunakan @


```python
import numpy as np

# Matriks A
A = np.array([[10, 7, 9],
              [9, 8, 10],
              [8, 9, 11]])
print("\nMatriks A :")
print(A)

# Matriks B
B = np.array([[1, 2, 3],
              [0, 3, 2],
              [-1, 4, 1]])
print("\nMatriks B :")
print(B)
```

```python
#Penjumlahan

C = A + B
print("\nMatriks C :")
print(C)

D = B + C
print("\nMatriks D :")
print(D)

#Pengurangan

E = D - C
print("\nMatriks E : ")
print(E)

F = A - E
print("\nMatriks F :")
print(F)

#Transpose

A_T = A.transpose()
print('\nMatriks A Transpose')
print(A_T)

B_T = B.transpose()
print('\nMatriks B Transpose')
print(B_T)

C_T = C.transpose()
print('\nMatriks C Transpose')
print(C_T)

D_T = D.transpose()
print('\nMatriks D Transpose')
print(D_T)

E_T = A.transpose()
print('\nMatriks E Transpose')
print(E_T)

F_T = F.transpose()
print('\nMatriks F Transpose')
print(F_T)

#perkalian
G = A @ B
print ("\nMatriks G :")
print(G)

# Fungsi untuk mengalikan dua matriks
def multiply_matrices(A, B):
    result = np.zeros((A.shape[0], B.shape[1]))
    for i in range(A.shape[0]):
        for j in range(B.shape[1]):
            for k in range(A.shape[1]):
                result[i][j] += A[i][k] * B[k][j]
    return result


# Mengalikan dua matriks tanpa menggunakan @
result = multiply_matrices(A, B)

# Menampilkan hasil
print("\nHasil perkalian matriks A dan B:")
print(result)
```
