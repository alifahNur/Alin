menghitung determinan matriks dan invers matriks 5 x 5
hitung invers matriks 5x5 menggunakan operasi gauss jordan

```python
import numpy as np

# Matriks input
A = np.array([
    [3, 1, 3],
    [2, 1, 2],
    [4, 2, 1]
])

# Kofaktor ekspansi dari baris kedua (baris indeks 1)
row_index = 1  # Baris kedua (indeks 1)
cofactor_expansion = sum(A[row_index, col_index] * np.linalg.det(np.delete(np.delete(A, row_index, axis=0), col_index, axis=1))
                        * (-1) ** (row_index + col_index) for col_index in range(A.shape[1]))

print("Kofaktor ekspansi dari baris kedua:", cofactor_expansion)
```
