#1.	Hitunglah PageRank dari graph berikut

```python
import numpy as np

# Jumlah node
N = 5

# Faktor redaman
d = 0.85

# Inisialisasi nilai PageRank
PR = np.array([1/N] * N)

# Representasi matriks adjacensi dari graf
# Baris mewakili link keluar dari node
# Kolom mewakili link masuk ke node
adj_matrix = np.array([
    [0, 0, 1, 0, 0],  # Node 1
    [1, 0, 0, 0, 0],  # Node 2
    [0, 1, 0, 0, 0],  # Node 3
    [1, 1, 0, 0, 1],  # Node 4
    [0, 1, 1, 1, 0]   # Node 5
])

# Jumlah link keluar
outgoing_counts = np.array([3, 3, 1, 2, 3])

# Jumlah iterasi maksimum
max_iterations = 20
tolerance = 1.0e-6

# Iterasi untuk menghitung nilai PageRank
for _ in range(max_iterations):
    new_PR = (1 - d) / N + d * adj_matrix.T.dot(PR / outgoing_counts)
    if np.linalg.norm(PR - new_PR) < tolerance:
        break
    PR = new_PR

# Tampilkan hasil PageRank
print("PageRank untuk setiap node:")
for i, pr in enumerate(PR, 1):
    print(f"Node {i}: {pr:.4f}")
```

#2. Carilah penyelesaian dari Sistem persamaan berikut  dengan Eliminasi Gauss

\[
\begin{align*}
x_1 - x_2 + x_3 &= 3 \\
2x_1 + x_2 + 3x_3 &= 8 \\
6x_1 - x_3 &= 4
\end{align*}
\]

```python
import numpy as np

# Matriks augmented dari sistem persamaan
A = np.array([
    [1, -1, 1],
    [2, 1, 3],
    [6, 0, -1]
], dtype=float)

b = np.array([3, 8, 4], dtype=float)

# Gabungkan matriks A dan vektor b menjadi matriks augmented
augmented_matrix = np.column_stack((A, b))

# Fungsi untuk melakukan eliminasi Gauss
def eliminasi_gauss(matriks_augmented):
    n = len(matriks_augmented)

    for i in range(n):
        # Pivoting: tukar baris jika elemen pivot adalah nol
        if matriks_augmented[i, i] == 0:
            for k in range(i+1, n):
                if matriks_augmented[k, i] != 0:
                    matriks_augmented[[i, k]] = matriks_augmented[[k, i]]
                    break

        # Mengeliminasi entri di bawah pivot
        for j in range(i+1, n):
            faktor = matriks_augmented[j, i] / matriks_augmented[i, i]
            matriks_augmented[j, i:] -= faktor * matriks_augmented[i, i:]

    # Substitusi balik
    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        x[i] = (matriks_augmented[i, -1] - np.sum(matriks_augmented[i, i+1:n] * x[i+1:n])) / matriks_augmented[i, i]

    return x

# Menyelesaikan sistem persamaan menggunakan eliminasi Gauss
solusi = eliminasi_gauss(augmented_matrix)

# Menampilkan hasil
print("Solusi dari sistem persamaan:")
for i, x in enumerate(solusi, 1):
    print(f"x{i} = {x:.4f}")
```
